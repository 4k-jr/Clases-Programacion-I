# Instrucciones básicas - C++

### Estructura general de un programa

```c++
#include <iostream>

using namespace std;

int main() {
    
    return 0;
}
```

`#include <iostream>` Declara los objetos que controlan la lectura y escritura en los flujos estándar. Esta inclusión suele ser el único encabezado que necesita incluir para realizar la entrada y salida de un programa de C++.

```c++
#include <iostream>
```

`using namespace std;` indica la presencia de este espacio de nombres, lo cual permite que todo lo que se encuentre en el nombre de espacio `std` esté disponible sin la necesidad de agregar el prefijo `std::` antes de utilizarlos.

```c++
using namespace std;
```

`int main()` es la función principal del programa, el cual será lo primero en ejecutarse al iniciar.

## Instrucciones básicas
### Salida de información

`cout` nos permite mostrar información en pantalla a través de la consola.

`endl` nos permite generar un salto de línea a la hora de mostrar información.

```c++
cout << "Hola mundo" << endl;
```

### Variables y tipos de datos

La sintaxis para declarar una variable es:

`tipo de dato` `nombre de la variable`

Por ejemplo:

```c++
int numero;
```

Es posible asignarle un valor a una variable, luego de haberla declarado, así cómo al momento de hacerlo.

por ejemplo:

```c++
numero = 10;

int numero2 = 5;
```

### Operadores

```c++
int suma = numero + numero2;
int resta = numero - numero2;
int multiplicacion = numero * numero2;
int division = numero / numero2;

cout << "La suma de " << numero << " + " << numero2 << " = " << suma << endl;
cout << "La resta de " << numero << " - " << numero2 << " = " << resta << endl;
cout << "La multiplicacion de " << numero << " x " << numero2 << " = " << multiplicacion << endl;
cout << "La division de " << numero << " / " << numero2 << " = " << division << endl;
```
