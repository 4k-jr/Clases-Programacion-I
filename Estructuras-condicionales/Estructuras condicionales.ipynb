# Estructuras condicionales

## Importar librerias de lectura y escritura

```c++
#include <iostream>
using namespace std;
```

## Estructura condicional simple

Ejecuta una determinada acción cuando se cumple una determinada condición

```c++
int numero = 5;

if(numero < 10){
    cout << "El valor alojado en la variable numero es menor que 10";
}
```

#### Explicación
Primeramente declaramos una variable de tipo entero llamado `numero`, a esta variable se le asigna un valor de 5.
posteriormente, se utiliza la instrucción `if` para evaluar la condición que se encuentra dentro de los paréntesis.

Cómo hemos visto en el salón de clases, tenemos diferentes operadores lógicos, en este caso estamos utilizando el operador `menor que` el cuál nos devuelve un valor `booleano`, es decir, verdadero o falso.

La estructura `if` va a hacer la evaluación, y en caso de que sea verdadero, va a ejecutar el segmento de código que se encuentre dentro de las llaves.

Al ser verdadero, vemos que se muestra el mensaje en pantalla que definimos, si la condición fuera falsa, por ejemplo, `numero > 10` el código no mostraría algún mensaje.

## Estructura condicional compuesta

Permite elegir entre dos alternativas posibles en función del cumplimiento o no de la condición a evaluar.

```c++
int numero2 = 10;

if(numero2 < 10){
    cout << "El valor alojado en la variable numero2 es menor que 10";
}else{
    cout << "El valor alojado en la variable numero2 es mayor o igual a 10";
}
```

#### Explicación
El ejemplo es similar al anterior, sin embargo, podemos encontrar algunas diferencias. En este caso hacemos la comparación con una variable entera llamada `numero2`, esta tiene un valor de 15, entonces al evaluar la condición `numero2 < 10` nos devuelve `falso`, y es de esta manera que utilizamos la instrucción `else` lo que nos permite agregar un caso contrario, si la condición evaluada es falsa.

Por ende, a diferencia del ejemplo anterior, se muestra un mensaje aun cuando la condición fue falsa.

## Estructura condicional anidada

Una estructura condicional es anidada cuando por la rama del `verdadero` o el `falso` de una estructura condicional hay otra estructura condicional.

Hay dos formas de trasladar este concepto a código las cuales son:

```c++
//Primera forma

if(numero2 < 10) {
    cout << "El valor alojado en la variable numero2 es menor que 10";
} else if(numero2 > 10) {
    cout << "El valor alojado en la variable numero2 es mayor que 10";
} else {
    cout << "El valor alojado en la variable numero2 es igual a 10";
}
```

```c++
//Segunda forma

if(numero2 < 10) {
    cout << "El valor alojado en la variable numero2 es menor que 10";
} else {
    if(numero2 > 10) {
        cout << "El valor alojado en la variable numero2 es mayor que 10";
    } else {
        cout << "El valor alojado en la variable numero2 es igual a 10";
    }
}
```

### Explicación
Al igual que el ejemplo anterior, lo que se lográ por medio de las estructuras condicionales anidadas, es poder tener más de 2 alternativas para la ejecución de códigos, lo cual va a depender directamente de la cantidad de condiciones que anidemos. En este ejemplo únicamente estamos utilizadon 2 condiciones, pero estas pueden ser más si es necesario.

Las dos formas de expresar estas estructuras son válidas, en caso de contar con más de 2 condiciones, se recomienda utilizar la primera forma, que nos brinda una mayor comodidad.

El resultado es igual en ambos casos, y vemos que de esta manera podemos obtener un mensaje más personalizado en comparación al ejemplo anterior, ya que si el valor de la variable `numero2` fuese mayor a 10, se mostraría el mensaje que le corresponde.
